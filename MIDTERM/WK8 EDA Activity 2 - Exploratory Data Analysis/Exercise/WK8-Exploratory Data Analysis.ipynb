```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"..\raw\chip-dataset-cleaned.csv",
                 delimiter=",")
df.info()
```

```python
df = df.drop(df.columns[[10,11,12]],axis=1)
df.head()
```

# N Chips Manufactured

```python
df_w = df.copy()
```

```python
chip_count = df_w['Vendor'].value_counts()
chip_count
```

```python
type(chip_count)
```

```python
chart = chip_count.plot(kind='bar',
                        color = ['#13538F','#13538F','#13538F','#A3CCE7','#A3CCE7'],
                        figsize=(10,5))
chart.bar_label(chart.containers[0], label_type='edge')
plt.title('Chips Manufactured (CPU+GPU)',
          fontfamily = 'Palatino Linotype',
          fontsize=12,
          fontweight='bold')
plt.xlabel('')
chart.set_xticklabels(chip_count.index, rotation=0)
plt.ylabel('N Chips')
plt.show()
```

# N CPU Manufactured

```python
df_cpu = df_w[df_w['Type']=='CPU']
df_cpu
```

```python
cpu_count = df_cpu['Vendor'].value_counts()
cpu_count
```

```python
chart = cpu_count.plot(kind='bar',figsize=(6,3), color = ['#0A1C5E','#4A99B0'])
plt.title('CPU Manufactured')
plt.ylabel('N Count')
chart.set_xticklabels(cpu_count.index, rotation=0)
plt.xlabel('')
plt.show()
```

# Lowest TDP (Thermal Design Power)

```python
df_tdp = df_w[['Product','TDP']]
df_tdp
```

```python
type(df_tdp)
```

```python
df_tdp = df_tdp.sort_values(by='TDP', ascending=True)
df_tdp
```

```python
df_tdp = df_tdp.head(7)
df_tdp
```

```python
plt.figure(figsize=(10,5))
plt.bar(df_tdp['Product'],df_tdp['TDP'])
plt.title('Products with the Lowest TDP')
plt.ylabel('Watts')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# GPU Process Size Trend

```python
df_ps_gpu = df_w[['Type','Date','Process_Size']].where(df_w['Type']=='GPU').dropna()
df_ps_gpu
```

```python
df_ps_gpu['Date'] = pd.to_datetime(df_ps_gpu['Date'])
df_ps_gpu['Year'] = df_ps_gpu['Date'].dt.year
df_ps_gpu 
```

```python
df_ps_gpu = df_ps_gpu.groupby('Year')['Process_Size'].mean().reset_index()
df_ps_gpu
```

```python
plt.figure(figsize=(10,5))
plt.plot(df_ps_gpu['Year'],df_ps_gpu['Process_Size'])
plt.title('GPU Process Size Trend')
plt.ylabel('nm')
plt.show()
```

# CPU Process Size Trend

## Processor Chip Production and Performance

The first chart highlights AMD's leadership in combined CPU and GPU production, showing its strong market presence in both categories. However, in the second chart, Intel surpasses AMD in CPU manufacturing, indicating its specialization and dominance in this category. In CPU-specific production, Intel takes the lead with over 1,200 units, reaffirming its position as a leader in this category, while AMD holds a strong second place. Overall, it showcases how these companies strategically position themselves to meet diverse market demands

## Thermal Design Power (TDP)
The dataset shows a wide range of Thermal Design Power (TDP) values, reflecting the different purposes of the chips. Low-TDP chips, like Intel's Atom series with just 1.0W, are designed for devices that need to save power and generate little heat, such as portable or embedded systems. On the other hand, high-TDP chips, like NVIDIA's Tesla series with TDPs up to 900W, focus on high performance for tasks like data centers and AI workloads. This range highlights how the industry balances energy efficiency for lightweight devices with powerful performance for demanding applications.


## GPU Process Size Trend

With the visualization of GPU process size and its historical trend, we can infer that the GPU process size decreases over time. This aligns with Moore's Law, which states that the number of transistors in a dense integrated circuit doubles approximately every two years, driving exponential growth in CPU and GPU capabilities. Additionally, this trend is supported by Huang's Law, which states that GPU performance will more than double every two years, further emphasizing the rapid advancements in GPU technology.
